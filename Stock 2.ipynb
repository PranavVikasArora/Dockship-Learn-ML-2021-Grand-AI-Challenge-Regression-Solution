```python
from datetime import datetime
stock2['Year'] = pd.to_datetime(stock2['Date']).dt.year
stock2['Month'] = pd.to_datetime(stock2['Date']).dt.month
stock2['date'] = pd.to_datetime(stock2['Date']).dt.day
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler1= MinMaxScaler()
scaler2 = MinMaxScaler()
```
